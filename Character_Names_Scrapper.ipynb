# Scrapping Characters Names from Wiki Fandom Page using Selenium 
- Data Manipulation and Cleaning

```python
import numpy as np 
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url = 'https://chaoswalking.fandom.com/wiki/Category:Characters'
driver.get(url)
```

```python
character_elements = driver.find_elements(By.CLASS_NAME,'category-page__member-link')
characters = []
for name in character_elements:
    characters.append(name.text)
characters    
```

```python
## Filter out elements that contain a category of characters
categories = list(filter(lambda x: 'Category' in x.text, character_elements))
categories
```

```python
## Collecting the href for each category
category_dict = {}
for category in categories:
    category_url = category.get_attribute('href')
    category_name = category.text
    category_dict[category_name] = category_url  
category_dict
```

```python
# Scraping each character in each category
for links in category_dict.keys():
    driver.get(category_dict[links])
    for name in driver.find_elements(By.CLASS_NAME,'category-page__member-link'):
        characters.append(name.text)
```

```python
characters
```

```python
characters_dict = {'Names' : characters}
characters_dict
```

```python
## Dropping duplicate names and removing categories
characters_df1 = pd.DataFrame(characters_dict)
characters_df1 = characters_df1[characters_df1['Names'].str.contains('Category') == False] 
characters_df1 = characters_df1.drop_duplicates()
characters_df1.replace('David Prentiss', 'Mayor')
```

```python
## Removing unwanted punctuation and splitting first names into seperate column
import re
characters_df1['Names'] = characters_df1['Names'].apply(lambda x: re.sub("[\(].*?[\)]", '', x))
characters_df1['First Name'] = characters_df1['Names'].apply(lambda x: x.split(' ')[0])
characters_df1['First Name'] = characters_df1['Names'].apply(lambda x: x if any(re.findall(r'Mr|Mrs|The', x)) else x.split()[0])

characters_df1
```

```python
## Exporting Dataframe into csv file
characters_df1.to_csv('characters.csv')
```
